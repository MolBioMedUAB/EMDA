```python
from EMDA import EMDA, __version__
print("EMDA version is:", __version__)

from numpy import average

import matplotlib.pyplot as plt

plt.style.use('ggplot')

from EMDA import ext_plot_contacts_frequencies_differences
```

```python
parameters = 'parameters.prmtop' # AMBER parameters file
trajectory = 'trajectory.nc'     # AMBER NetCDF coordinates (10 frames)
```

```python
emda1 = EMDA(parameters=parameters, trajectory=trajectory)
emda2 = EMDA(parameters=parameters, trajectory=trajectory)
```

```python
def run_measures(emda, first_frame, last_frame):
    emda.select('protein', 'protein')
    emda.add_contacts('contacts_prot', 'protein', sel_env=1.5)
    emda.run(start=first_frame, end=last_frame)

    emda.analyse_contacts_frequency('contacts_prot_freq', 'contacts_prot')


    return emda

```

```python
emda1 = run_measures(emda1, 1, 5)
emda2 = run_measures(emda1, 6, 10)
```

```python
emda1.analyses['contacts_prot_freq'].result
```

```python
emda2.analyses['contacts_prot_freq'].result
```

```python
res = ext_plot_contacts_frequencies_differences(emda1.analyses['contacts_prot_freq'].result, emda2.analyses['contacts_prot_freq'].result, return_results=True, threshold=0)
```
